# Homework Assignment: Exploratory Data Analysis (EDA)
In this assignment, you will apply the EDA techniques you've learned to a new dataset. Complete the tasks below to practice data loading, inspection, and visualization.

**Dataset:** The 'Iris' dataset from the UCI Machine Learning Repository is preloaded for you.

```python
# Step 1: Load and Inspect the Data
import pandas as pd

# Load the dataset
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'
data = pd.read_csv(url)
data.head()
```

### Task 1: Basic Data Inspection
- Display the first 5 rows of the dataset.
- Show the data types of each column.
- Display summary statistics.

```python
# Your code here
print(data.head(5))
print("\n")
print(data.dtypes)
print("\n")
print(data.describe(include='all').T)
```

### Task 2: Check for Missing Values
- Check if there are any missing values in the dataset.

```python
# Your code here
data.isnull().sum()
```

### Task 3: Univariate Analysis
- Plot the distribution of the `sepal_length` column.
- Count the unique values in the `species` column and plot a bar chart.

```python
import matplotlib.pyplot as plt
import seaborn as sns

data['sepal_length'].hist(figsize=(8, 6), bins=20)
plt.title('Histogram of Sepal Length')
plt.tight_layout()
plt.show()

species_counts = data['species'].value_counts()
species_counts.plot(kind='bar', color='green', figsize=(8, 6))
plt.title('Count of Unique Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

### Task 4: Bivariate Analysis
- Create a scatter plot of `sepal_length` vs `sepal_width` colored by species.
- Generate a pairplot for the dataset using Seaborn.

```python
sns.scatterplot(data=data, x='sepal_length', y='sepal_width', hue='species', palette='viridis')
plt.title('Sepal Length vs Sepal Width by Species')
plt.tight_layout()

sns.pairplot(data, hue='species', palette='viridis')
plt.show()
```

### Task 5: Grouping and Aggregation
- Calculate the average `sepal_length` for each species.
- Find the maximum `petal_width` for each species.

```python
print(data.groupby('species')['sepal_length'].mean())
print("\n")
print(data.groupby('species')['petal_width'].max())
```

### Task 6: Visualizations
- Create a histogram for `petal_length`.
- Generate a box plot for `petal_width` by species.

```python
# Your code here
data['petal_length'].hist(figsize=(8, 6), bins=20, color='blue')
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

sns.boxplot(data=data, x='species', y='petal_width', palette='pastel', hue='species', legend=False)
plt.title('Box Plot of Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width')
plt.tight_layout()
plt.show()
```
