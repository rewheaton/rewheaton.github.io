# Neural Networks with CIFAR-10 - Student Activity

The CIFAR-10 dataset (short for the Canadian Institute For Advanced Research) is a popular dataset for machine learning and deep learning, primarily used for image classification tasks. It consists of 60,000 color images categorized into 10 different classes. Each image is 32x32 pixels in size and has 3 color channels (RGB).

## Step 1: Load and Preprocess the Dataset

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import cifar10
import matplotlib.pyplot as plt

# Load CIFAR-10 Dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Normalize pixel values to [0, 1]
X_train, X_test = X_train / 255.0, X_test / 255.0

print("Training data shape:", X_train.shape)
print("Test data shape:", X_test.shape)
```

## Step 2: Visualize the Dataset

```python
# TODO: Visualize the first 10 images from the training dataset
import matplotlib.pyplot as plt
class_names = ['Airplane', 'Automobile', 'Bird', 'Cat', 'Deer', 'Dog', 'Frog', 'Horse', 'Ship', 'Truck']
# Hint: Use a for loop and plt.subplot

for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(X_train[i])
    plt.title(class_names[y_train[i][0]])
    plt.axis('off')
```

## Step 3: Define the Neural Network

```python
# TODO: Define a neural network with Flatten, Dense layers, and the softmax activation function
model = Sequential([
    Flatten(input_shape=(32, 32, 3)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])
```

## Step 4: Compile the Model

```python
# TODO: Compile the model using the Adam optimizer, sparse categorical crossentropy loss, and accuracy as a metric
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

## Step 5: Train the Model

```python
# TODO: Train the model for 10 epochs using 20% of the training data for validation
history = model.fit(X_train, y_train, epochs=10, validation_split=0.2)
```

## Step 6: Evaluate the Model

```python
# TODO: Evaluate the model on the test data and print the test accuracy
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
```

## Step 7: Visualize Training Results

```python
# TODO: Plot the training and validation accuracy over epochs
history = model.history.history
plt.plot(history['accuracy'], label='Training Accuracy')
plt.plot(history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
```
