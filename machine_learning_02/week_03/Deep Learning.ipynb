## Step 1: Load and Explore the Dataset

```python
# TODO: Load the dataset and normalize it
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
# Hint: Use fashion_mnist.load_data() and normalize by dividing by 255.0
# Display the shapes of the training and test datasets

fashion_data = fashion_mnist.load_data()
# Unpack training and test data
(train_images, train_labels), (test_images, test_labels) = fashion_data

# Normalize the pixel values
train_images = train_images / 255.0
test_images = test_images / 255.0

# Display shapes
print(f"Training images shape: {train_images.shape}")
print(f"Training labels shape: {train_labels.shape}")
print(f"Test images shape: {test_images.shape}")
print(f"Test labels shape: {test_labels.shape}")


# Get unique labels from training data
unique_labels = sorted(set(train_labels))
print("Unique labels in the training set:", unique_labels)
```

## Step 2: Visualize the Data

```python
# TODO: Visualize the first 10 images in the training dataset with their labels
import matplotlib.pyplot as plt
# Hint: Use plt.imshow() and class_names for displaying labels

#  Source: https://www.tensorflow.org/datasets/catalog/fashion_mnist
class_names = {
    0: 'T-shirt/top',
    1: 'Trouser',
    2: 'Pullover',
    3: 'Dress',
    4: 'Coat',
    5: 'Sandal',
    6: 'Shirt',
    7: 'Sneaker',
    8: 'Bag',
    9: 'Ankle boot'
}

plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(train_images[i], cmap='gray')
    plt.title(class_names[train_labels[i]])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

## Step 3: Build the Neural Network

```python
# TODO: Define the model with Flatten, Dense layers, and softmax activation
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
# Hint: Use Sequential API

model = Sequential([
    Flatten(input_shape=(28, 28)), 
    Dense(128, activation='relu'),   
    Dense(64, activation='relu'),    
    Dense(10, activation='softmax')  
])

# Display model summary
model.summary()
```

## Step 4: Compile the Model

```python
# TODO: Compile the model with Adam optimizer and sparse categorical crossentropy loss
# Hint: Use model.compile()
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

## Step 5: Train the Model

```python
# TODO: Train the model for 10 epochs and include validation data
# Hint: Use model.fit()
history = model.fit(train_images, train_labels, 
                    epochs=10, 
                    validation_data=(test_images, test_labels))
```

## Step 6: Evaluate the Model

```python
# TODO: Evaluate the model on the test dataset
# Hint: Use model.evaluate()
test_loss, test_accuracy = model.evaluate(test_images, test_labels)
print(f"Accuracy: {test_accuracy:.4f}")
```

## Step 7: Visualize Training Results

```python
# TODO: Plot training and validation accuracy over epochs
# Hint: Use history.history['accuracy'] and history.history['val_accuracy']

plt.figure(figsize=(10, 5))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
