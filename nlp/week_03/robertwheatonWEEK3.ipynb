# N-grams

<b>N-grams</b> are a contiguous sequence of <i>n</i> items from a given sample of text or speech.

<b>Examples:</b>

- “Latest discovery” is a 2-gram.

- “Advances in NLP” is a 3-gram.

The NLTK library offers methods to explore n-grams.

Extract 4-grams from the piece of text below:

```python
import nltk
import pandas as pd 
text= "Sweden is part of the geographical area of Fennoscandia. The climate is in \
general mild for its northerly latitude due to significant maritime influence. \
In spite of the high latitude, Sweden often has warm continental summers, \
being located in between the North Atlantic, the Baltic Sea, and vast Russia. \
The general climate and environment vary significantly from the south and north \
due to the vast latitudal difference, and much of Sweden has reliably cold and \
snowy winters. Southern Sweden is predominantly agricultural, while the north is \
heavily forested and includes a portion of the Scandinavian Mountains."

words = text.split(' ')

# Code to extract 4-grams:
(pd.Series(nltk.ngrams(words,4)).value_counts())[:10]
```

## Stopwords 

Stopwords are the English words add little or no meaning to a sentence. They can generally be ignored without sacrificing the meaning of a sentence. 

Examples of Stopwords: 
- like
- the
- he
- have

Reference: https://www.tutorialspoint.com/python_text_processing/python_remove_stopwords.htm

```python
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
nltk.download('stopwords')

# Function to plot stopwords:
def plot_top_stopwords_barchart(text):
    stop=set(stopwords.words('english'))
    
    new= text.str.split()
    new=new.values.tolist()
    corpus=[word for i in new for word in i]
    from collections import defaultdict
    dic=defaultdict(int)
    for word in corpus:
        if word in stop:
            dic[word]+=1
            
    top=sorted(dic.items(), key=lambda x:x[1],reverse=True)[:10] 
    x,y=zip(*top)
    plt.bar(x,y)
```

Plot the top 10 stopwords using this function.

```python
#your code here
plot_top_stopwords_barchart(news['headline_text'])
```

# Visualize n-grams with Bar Plots.

Understanding common phrases found in text helps clarify context.

The following code defines a function that identifies n-grams in a text.

```python
from sklearn.feature_extraction.text import CountVectorizer #scikit learn library: https://scikit-learn.org/stable/
import seaborn as sns # seaborn is a visualization library; more here: https://seaborn.pydata.org/introduction.html#:~:text=Seaborn%20is%20a%20library%20for,explore%20and%20understand%20your%20data.
def get_top_ngram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(n, n)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) 
                  for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:10]
```

### Use the function with Seaborn's barplot function to plot the top n-grams.

```python
bigrams=get_top_ngram(news['headline_text'],2)[:10] 
x,y=map(list,zip(*bigrams)) 
sns.barplot(x=y,y=x)
```

```python
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer 
from nltk import word_tokenize 
nltk.download('wordnet')
def preprocess_news(df):
    corpus=[]
    stem=PorterStemmer()
    lem=WordNetLemmatizer()
    stop=set(stopwords.words('english'))
    for news in df['headline_text']:
        words=[w for w in word_tokenize(news) if (w not in stop)]
        
        words=[lem.lemmatize(w) for w in words if len(w)>2]
        
        corpus.append(words)
    return corpus

#here is the list wordlist
wordlist=preprocess_news(news)
```

Generate the wordcloud for the headlines.

```python
wc = WordCloud(width=1500, 
               height=1000, 
               random_state=1, 
               background_color='white', 
               collocations= False, 
               stopwords=STOPWORDS).generate(str(wordlist))
plt.figure(figsize=(15,10))
plt.imshow(wc)
```
